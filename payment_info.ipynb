# Extract Data

```python
#dependency
import pandas as pd
from sqlalchemy import create_engine
```

```python
#import and read csv data into pandas dataframe for Integra Insurance
integra_data=("Resources/integra_insurance.csv")
#create data frame
integra_data_df=pd.read_csv(integra_data)
#to see the data
integra_data_df.head()
```

```python
#clean data
#pick only usefull column data
integra_data_df1=integra_data_df[["LineNumber","Name","ServiceEndDate","PaidAmount"]].copy()
integra_data_df1.head()
```

```python
#rename the columns
integra_data_df1_re=integra_data_df1.rename(columns={"LineNumber":"id",
                                                    "Name":"patient_name",
                                                     "ServiceEndDate":"payment_date",
                                                     "PaidAmount":"total_paid"
                                                        })
#to see the data
integra_data_df1_re.head()            
```

```python
#set primary key as patient_id
integra_data_df1_re.set_index("id", inplace=True)
#to see the data
integra_data_df1_re.head()
```

```python
#import and read csv data into pandas dataframe for VNSNY Insurance
vnsny_data=("Resources/vnsny_insurance.csv")
#create data frame
vnsny_data_df=pd.read_csv(vnsny_data)
vnsny_data_df.head()
```

```python
#clean the data
#pick only usefull column data
vnsny_data_df1=vnsny_data_df[["LineNumber","Patient Name","Payment Date","Total Pay"]].copy()
vnsny_data_df1
```

```python
#rename the columns
vnsny_data_df1_re=vnsny_data_df1.rename(columns={
    "LineNumber":"id",
    "Patient Name":"patient_name",
    "Payment Date":"payment_date",
    "Total Pay":"total_paid"   
})
#to see the data
vnsny_data_df1_re.head()
```

```python
#set the primary key as patient_id
vnsny_data_df1_re.set_index("id", inplace = True)
#to see the data
vnsny_data_df1_re.head()
```

# Create a  connection 

```python
#create database connection
connection_string="Aakash:Password@localhost:5432/insurance_db"
engine=create_engine(f"postgresql://{connection_string}")
```

```python
#confirm the tables
engine.table_names()
```

```python
#load data frame into database
integra_data_df1_re.to_sql(name="integra_insurance", con=engine,if_exists="append", index=True)
```

```python
#load data frame into database
vnsny_data_df1_re.to_sql(name="vnsny_insurance", con=engine, if_exists="append", index=True)
```

```python
#see the integra_insurance table data  
member_integra = pd.read_sql_query("select * from integra_insurance", con=engine)
member_integra.head()
```

```python
#see the vnsny_insurance table data  
member_vnsny=pd.read_sql_query("select * from vnsny_insurance", con=engine)
member_vnsny.head()
```

```python
#to save the query data for Integra Insurance
member_integra.to_csv("output\integra.csv")
```

```python
#to save the query data for VNSNY Insurance
member_vnsny.to_csv("output\vnsny.csv")
```
